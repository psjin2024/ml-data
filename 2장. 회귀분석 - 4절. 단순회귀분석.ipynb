```python
X = [32,64,96,118,126,144,152,158]
Y = [17,24,62,49,52,105,130,125]
```

```python
import numpy as np
A=np.c_[X, np.ones(8)]  # np.c_ : 열방향으로 배열을 합침
B=np.array(Y)
A
```

```python
B
```

```python
model = np.linalg.inv(A.T @ A) @ A.T @ B
model
```

```python
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(X, Y)    # 행렬의 곱계산  linregress()
```

```python
print(f"slope: {slope}")
print(f"intercept: {intercept}")
print(f"r_value: {r_value}")
print(f"p_value: {p_value}")
print(f"std_err: {std_err}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib inline
# %config InlineBackend.figure_format = 'retina'  
```

```python
plt.scatter(X, Y)
plt.plot(X, slope*np.array(X) + intercept, '-')
plt.show()
```

```python
fit = np.polyfit(X, Y, 1)
fit
```

```python
fit_fn = np.poly1d(fit) # 항의 계수를 이용해서 함수를 정의
fit_fn
```

```python
plt.scatter(X, Y)
```

```python
plt.scatter(X, Y)
plt.plot(X, fit_fn(X))
plt.show()
```

```python
fit = np.polyfit(X, Y, 2)
fit_fn = np.poly1d(fit)
sample_X = np.linspace(30,160,100)
plt.scatter(X, Y)
plt.plot(sample_X, fit_fn(sample_X))
plt.show()
```

```python
def draw_polyfit(X, Y, deg=1):
    import numpy as np
    fit = np.polyfit(X, Y, deg)
    print(fit)
    fit_fn = np.poly1d(fit)
    sample_X = np.linspace(min(X), max(X), 100)
    plt.scatter(X, Y)
    plt.plot(sample_X, fit_fn(sample_X))
    plt.show()
```

```python
draw_polyfit(X, Y, 3)
```

```python
draw_polyfit(X, Y, 4)
```

```python
draw_polyfit(X, Y, 5)
```

```python
draw_polyfit(X, Y, 6)
```
