```python
import pandas as pd   # 분산, 표준편차 
import numpy as np  # 모분산, 모 표준편차
```

```python
data = [1,2,3,4, 5]
print("모표준편차", np.std(data, ddof=0))   #std  모표준편차
print("표본표준편차", np.std(data, ddof=1))   #std  

data_df = pd.DataFrame(data)
print("표본표준편차",data_df.std(ddof=1))   # 표본표준편차  일반적으로 많이 쓰임
print("모표준편차",data_df.std(ddof=0))  #std    델타자유도=ddof   보고시 명확히 해야할 필요가 있음
```
